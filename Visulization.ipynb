```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(17, 5))

X_Axis = [28, 72]
Y_Axis = ["Incorrect Output", "Correct Output"]
ax[1].pie(X_Axis, labels=Y_Axis, autopct="%0.1f%%")
ax[1].set_title("BERT")

X_Axis = [72, 12, 16]
Y_Axis = ["Accurate Output", "Type-2 Error", "Type-1 Error"]
ax[2].pie(X_Axis, labels=Y_Axis, autopct="%0.1f%%", explode=[0.1, 0.3, 0])
ax[2].set_title("BERT")

X_Axis = ["Accuracy", "Recall", "Precision"]
Y_Axis = [72, 79, 74]

ax[0].bar(X_Axis, Y_Axis, width=0.6)
ax[0].set_title("BERT")  # Use set_title, not title
ax[0].set_ylabel("Percentage")  # Use set_ylabel, not ylabel
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(17, 5))

# First Subplot
x_axis = [20, 80]
y_axis = ["Incorrect Output", "Correct Output"]
ax[1].pie(x_axis, labels=y_axis, autopct="%0.1f%%")
ax[1].set_title("T-5")

# Second Subplot
x_axis = [4, 16, 80]
y_axis = ["Type-2 Error", "Type-1 Error", "Accurate Output"]
ax[2].pie(x_axis, labels=y_axis, autopct="%0.1f%%", explode=[0.1, 0.3, 0])
ax[2].set_title("T-5")

# Third Subplot
x_axis = ["Accuracy", "Recall", "Precision"]
y_axis = [80, 93, 77]

ax[0].bar(x_axis, y_axis, width=0.6, color="red")
ax[0].set_title("T-5")
ax[0].set_ylabel("Percentage")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the first subplot (RoBERTa)
data_roberta = [30, 70]
labels_roberta = ["Incorrect Output", "Correct Output"]

# Data for the second subplot (RoBERTa)
data_roberta_errors = [8, 12, 80]
labels_roberta_errors = ["Type-2 Error", "Type-1 Error", "Accurate Output"]

# Data for the third subplot (RoBERTa)
data_roberta_metrics = [85, 90, 78]
labels_roberta_metrics = ["Accuracy", "Recall", "Precision"]

# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(17, 5))

# First Subplot (Pie Chart)
ax[0].pie(data_roberta, labels=labels_roberta, autopct="%0.1f%%")
ax[0].set_title("RoBERTa")

# Second Subplot (Pie Chart)
ax[1].pie(data_roberta_errors, labels=labels_roberta_errors, autopct="%0.1f%%", explode=[0.1, 0.3, 0])
ax[1].set_title("RoBERTa")

# Third Subplot (Bar Chart)
ax[2].bar(labels_roberta_metrics, data_roberta_metrics, width=0.6, color="blue")
ax[2].set_title("RoBERTa")
ax[2].set_ylabel("Percentage")

# Display the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

X_Axis = ["Accuracy", "Recall", "Precision"]
BERT = [72, 79.31, 74.19]
T_5 = [80, 93.1, 77.14]
RoBERTa = [85, 90.5, 78.9]  # Add RoBERTa data here

plt.plot(X_Axis, BERT, color="green", label="BERT")
plt.plot(X_Axis, T_5, color="red", label="T-5")
plt.plot(X_Axis, RoBERTa, color="blue", label="RoBERTa")  # Add RoBERTa data here

plt.title("BERT vs T-5 vs RoBERTa")
plt.ylabel("Percentage")
plt.legend()

plt.grid()
plt.show()
```
